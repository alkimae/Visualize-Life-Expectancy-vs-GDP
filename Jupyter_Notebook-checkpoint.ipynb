#### Import Python Modules

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

#### Import CSV & View Data Header

```python
df = pd.read_csv("all_data.csv")
df.head()
```

#### Rename Life Excpectancy Column

```python
df = df.rename({"Life expectancy at birth (years)":"Life Expectancy"}, axis = "columns")
df.head()
```

#### Explore the Data

```python
print(df.Country.unique())
```

6 Countries represented

```python
print(df.Year.unique())
```

Year range: 2000 - 2015

### Question 1:
#### Has life expectancy shown a consistent increase over time in the six nations?

As seen in the charts below, life expectancy has consistently increased over time in all six nations, with Zimbabwe experiencing the most substantial improvement.

```python
#plot life expectancy per year across all countries
sns.lineplot(x='Year', y='Life Expectancy', hue='Country', data=df, palette='tab10')

plt.legend(title='Country', bbox_to_anchor=(1.05, 1))
plt.title('Life Expectancy per Year Across 6 Countries')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
#plot life expectancy per year for each country
sns.set(style="whitegrid")

countries = df['Country'].unique()

num_rows = (len(countries) + 1) // 2
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8), sharex=True)
axes = axes.flatten()

for i, country in enumerate(countries):
    subset = df[df['Country'] == country]
    sns.lineplot(x='Year', y='Life Expectancy', data=subset, ax=axes[i])
    axes[i].set_title(country)

plt.tight_layout()
plt.show()
```

### Question 2:
#### Has the Gross Domestic Product (GDP) exhibited a steady rise over time in these nations?

As seen in the charts below, Gross Domestic Product (GDP) shows a consistent upward trajectory for all countries in the study, with China leading in significant economic growth.

```python
#plot GDP per year across all countries
sns.lineplot(x='Year', y='GDP', hue='Country', data=df, palette='tab10')

plt.legend(title='Country', bbox_to_anchor=(1.05, 1))
plt.title('GDP per Year Across 6 Countries')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```

```python
#plot GDP per year for each country
sns.set(style="whitegrid")

countries = df['Country'].unique()

num_rows = (len(countries) + 1) // 2
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8), sharex=True)
axes = axes.flatten()

for i, country in enumerate(countries):
    subset = df[df['Country'] == country]
    sns.lineplot(x='Year', y='GDP', data=subset, ax=axes[i])
    axes[i].set_title(country)

plt.tight_layout()
plt.show()
```

### Question 3:

#### Is there a discernible correlation between a country's GDP and its life expectancy?

As seen in the following scatterplot, a positive correlation exists between a country's GDP and its life expectancy within the selected nations.

```python
# Plot life expectancy vs GDP per country
country_palette = sns.color_palette("husl", n_colors=len(df['Country'].unique()))

plt.figure(figsize=(10, 6))
for country, color in zip(df['Country'].unique(), country_palette):
    country_data = df[df['Country'] == country]
    plt.scatter(country_data['Life Expectancy'], country_data['GDP'], label=country, color=color)

plt.title('Life Expectancy vs. GDP')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')
plt.legend()
plt.show()
```

### Question 4:
#### What is the average life expectancy across the selected nations?

As seen in the figure below, the average life expectancy ranges from the mid to high 70s for most countries, except for Zimbabwe, where it is notably lower at 50.

```python
#plot average life expectancy per country
avg_gdp_per_country = df.groupby('Country')['Life Expectancy'].mean().reset_index()

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Life Expectancy', data=avg_gdp_per_country)

plt.title('Average Life Expectancy per Country')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy')
plt.show()
```

```python
#plot average GDP per country
avg_gdp_per_country = df.groupby('Country')['GDP'].mean().reset_index()

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='GDP', data=avg_gdp_per_country)

plt.title('Average GDP per Country')
plt.xlabel('Country')
plt.ylabel('Average GDP')
plt.show()
```

### Question 5:
#### How is life expectancy distributed within these countries?

As seen in the histogram below, the distribution of life expectancy exhibits a left skew, indicating that the majority of observations are concentrated on the right side of the distribution.

```python
#plot histogram of life expectancy distribution
plt.hist(df['Life Expectancy'])

plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy at Birth in Years')
plt.show()
```

```python
#plot histogram of GDP distribution
plt.hist(df['GDP'])

plt.title('Distribution of GDP')
plt.xlabel('GDP in Trillions of Dollars')
plt.show()
```
